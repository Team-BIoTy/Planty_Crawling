## 작형 및 재배관리

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
# chrome options
options = Options()
options.add_argument("--headless") # 백그라운드로 실행

# webdriver 설정
service = Service(executable_path=ChromeDriverManager().install()) 
driver = webdriver.Chrome(service=service, options=options)

# 크롤링할 사이트 URL
url = "https://www.nongsaro.go.kr/portal/ps/psz/psza/contentMain.ps?menuId=PS03168"
driver.get(url)

def get_element(xpath):
    # 제목 (h4) 가져오기 (단일 요소)
    subtitle_elem = driver.find_element(By.XPATH, xpath + '/h4')
    subtitle_text = subtitle_elem.text if subtitle_elem else "제목 없음"

    # 리스트(ul/li) 항목 가져오기 (여러 개 요소)
    content_elems = driver.find_elements(By.XPATH, xpath + '/ul/li')
    content_texts = [elem.text for elem in content_elems]  # 리스트의 각 요소에서 .text 추출

    # HTML 생성
    html_content = f"<h4>{subtitle_text}</h4>"
    for text in content_texts:
        html_content += f"<p>{text}</p>"

    html_content += "<hr>"
    return html_content

# XPath를 이용해 요소 가져오기
try:
    # HTML 콘텐츠 생성
    html_content = "<html><body><h1>작형및재배관리</h1>"

    for i in range(1, 4):
        html_content += get_element(f'//*[@id="printZone"]/div/div[2]/div[{i}]')
    
    html_content += "</body></html>"

    # HTML 파일로 저장
    with open("../data/농사로_작형및재배관리.html", "w", encoding="utf-8") as file:
        file.write(html_content)

    print("HTML 파일이 생성되었습니다.")

except Exception as e:
    print("Error:", e)

# 드라이버 종료
driver.quit()
```

## 도시농업

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
# chrome options
options = Options()
options.add_argument("--headless") # 백그라운드로 실행

# webdriver 설정
service = Service(executable_path=ChromeDriverManager().install()) 
driver = webdriver.Chrome(service=service, options=options)

# 크롤링할 사이트 URL
url = "https://www.yongin.go.kr/home/atc/cityConsum/cityConsum03/cityConsum03_01/cityConsum03_01_01.jsp"
driver.get(url)

# XPath를 이용해 요소 가져오기
try:
    
    # HTML 콘텐츠 생성
    html_content = "<html><body><h1>도시농업</h1>"
    
    # 탭 이동동
    for i in range(1, 5):
        if i != 1:
            driver.find_element(By.XPATH, f'//*[@id="contents"]/div[3]/div[1]/ul/li[{i}]/a').click()
        
        # 소제목 추출
        headings = driver.find_elements(By.XPATH, '//*[@id="contents"]/div[2]/h3')
        for heading in headings:
            html_content += f"<h3>{heading.text}</h3>"

        # 내용 추출
        elements = driver.find_elements(By.XPATH, '//*[@id="contents"]/div[3]/*[not(contains(@class, "tab_list"))]')
        for element in elements:
            html_content += f"<p>{element.text}</p>"
        html_content += '<hr>'

    html_content += "</body></html>"

    # HTML 파일로 저장
    with open("../data/농업기술센터_도시농업.html", "w", encoding="utf-8") as file:
        file.write(html_content)

    print("HTML 파일이 생성되었습니다.")

except Exception as e:
    print("Error:", e)

# 드라이버 종료
driver.quit()
```

## 도시농업 정보

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Chrome options 설정
options = Options()
options.add_argument("--headless")  # 백그라운드 실행

# WebDriver 설정
service = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 크롤링할 사이트 URL
url = "https://www.modunong.or.kr:449/home/kor/guide/ready/index.do"
driver.get(url)

# XPath를 이용해 요소 가져오기
try:
    # HTML 콘텐츠 생성
    html_content = "<html><body><h1>나만의 텃밭 가꾸기</h1>"

    # 탭 이동
    for i in range(1, 6):
        if i != 1:
            tab_button = driver.find_element(By.XPATH, f'//*[@id="cmmnForm"]/div/div[1]/ul/li[{i}]/button')
            tab_button.click()
            time.sleep(1)  # 클릭 후 페이지가 변경될 시간을 줌

        # 소제목 추출
        heading_steps = driver.find_elements(By.XPATH, f'//*[@id="cmmnForm"]/div/div[1]/ul/li[{i}]/button/div')
        heading_texts = driver.find_elements(By.XPATH, f'//*[@id="cmmnForm"]/div/div[1]/ul/li[{i}]/button/span')

        for step, text in zip(heading_steps, heading_texts):
            html_content += f"<h3>{step.text} {text.text}</h3>"

        # 내용 추출
        elements = driver.find_elements(By.XPATH, '//*[@id="cmmnForm"]/div/*[not(contains(@class, "step_list_box"))]')
        for element in elements:
            html_content += f"<p>{element.text}</p>"
        html_content += "<hr>"

    html_content += "</body></html>"

    # HTML 파일로 저장
    with open("../data/모두가도시농부_나만의텃밭가꾸기.html", "w", encoding="utf-8") as file:
        file.write(html_content)

    print("HTML 파일이 생성되었습니다.")

except Exception as e:
    print("Error:", e)

# 드라이버 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Chrome options 설정
options = Options()
# options.add_argument("--headless")  # 백그라운드 실행

# WebDriver 설정
service = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 크롤링할 사이트 URL
url = "https://www.modunong.or.kr:449/home/kor/garden/type/index.do"
driver.get(url)

# XPath를 이용해 요소 가져오기
try:
    # HTML 콘텐츠 생성
    html_content = "<html><body><h1>텃밭유형</h1>"

    # 탭 이동
    for i in range(1, 26):
        if i % 8 == 1:
            page = i // 8 + 1
            driver.find_element(By.XPATH, f'//*[@id="paging"]/a[{page}]').click()
        tab_button = driver.find_element(By.XPATH, f'//*[@id="tab02"]/div[2]/div[2]/ul/li[{i%8}]/div/a/div[2]/div/img').click()
        time.sleep(1)

        # 소제목 추출
        heading = driver.find_elements(By.XPATH, f'//*[@id="cmmnForm"]/div[1]/div[1]/strong/h2')
        for head in heading:
            html_content += f"<h2>{head.text}</h2>"

        # 내용 추출
        elements = driver.find_elements(By.XPATH, '//*[@id="cmmnForm"]/div[1]/div[2]')
        for element in elements:
            html_content += f"<p>{element.text}</p>"
        html_content += "<hr>"

        driver.back()
    html_content += "</body></html>"

    # HTML 파일로 저장
    with open("../data/모두가도시농부_텃밭유형.html", "w", encoding="utf-8") as file:
        file.write(html_content)

    print("HTML 파일이 생성되었습니다.")

except Exception as e:
    print("Error:", e)

# 드라이버 종료
driver.quit()
```
