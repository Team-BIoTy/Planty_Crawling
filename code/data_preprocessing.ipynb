```python
import pandas as pd

# 잎을 관상하는 실내식물
leaf_info_df = pd.read_csv("../data/leaf_info.csv")
leaf_detail_df = pd.read_csv("../data/leaf_detail.csv")
leaf_manage_df = pd.read_csv("../data/leaf_manage.csv")

# 건조에 강한 실내식물
dry_info_df = pd.read_csv("../data/dry_info.csv")
dry_manage_df = pd.read_csv("../data/dry_manage.csv")
```

```python
# '이름'을 기준으로 병합
leaf = leaf_info_df.merge(leaf_detail_df, on='이름', how='outer')
leaf = leaf.merge(leaf_manage_df, on='이름', how='outer')

# 결과 출력
leaf.head()
```

```python
leaf.columns
```

```python
leaf.to_csv("../data/leaf.csv", index=False)
```

```python
# '이름'을 기준으로 병합
dry = dry_info_df.merge(dry_manage_df, on='이름', how='outer')
dry.head()
```
